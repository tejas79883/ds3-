```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('Titanic.csv')
dataset.head()
```

```python
#Distribution plot
#distplot Dist plot gives us the histogram of the selected continuous variable
import seaborn as sns
sns.distplot(x = dataset['age'], bins = 10)
```

```python
sns.distplot(dataset['age'], bins = 10,kde=False)
```

```python
#join plot combination of the distplot of two variables.
import seaborn as sns
# For Plot 1
sns.jointplot(x = dataset['age'], y = dataset['fare'], kind =
'scatter')

```

```python
# For Plot 2
sns.jointplot(x = dataset['age'], y = dataset['fare'], kind = 'hex')
```

```python
#Rug plot used to draw small bars along the x-axis for each point in the dataset
sns.rugplot(dataset['fare'])
```

```python
#categorical plot
#bar plot sed to display the mean value for each value in a categorical column, against a
#numeric columns.
sns.barplot(x='sex', y='age', data=dataset)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.barplot(x='sex', y='age', data=dataset, estimator=np.std)
```

```python
#count plotThe count plot is similar to the bar plot, however it displays the count of the categories in a
#specific column
sns.countplot(x='sex', data=dataset)
```

```python
#Box plot display the distribution of the categorical data in the form of quartiles.
#The centre of the box shows the median value.
sns.boxplot(x='sex', y='age', data=dataset)
```

```python
sns.boxplot(x='sex', y='age', data=dataset, hue="survived")
```

```python
#viloin plot
#used to plot
#the violin plot. Like the box plot, the first parameter is the categorical column, the second
#parameter is the numeric column while the third parameter is the dataset
sns.violinplot(x='sex', y='age', data=dataset)
```

```python
sns.violinplot(x='sex', y='age', data=dataset, hue='survived')
```

```python
#Advance plot 
#strip plot draws a scatter plot where one of the variables is categorical
sns.stripplot(x='sex', y='age', data=dataset, jitter=False)
```

```python
sns.stripplot(x='sex', y='age', data=dataset, jitter=True)
```

```python
sns.stripplot(x='sex', y='age', data=dataset, jitter=True, hue='survived')
```

```python
#swarm plot the points
#are adjusted in such a way that they don't overlap
sns.swarmplot(x='sex', y='age', data=dataset)
```

```python
sns.swarmplot(x='sex', y='age', data=dataset, hue='survived')
```

```python
#matrix plot
#heat map used to plot correlation between numeric columns in the form of a
#matrix.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
dataset = pd.read_csv('Titanic.csv')
dataset.head()
```

```python
# Assuming dataset is your DataFrame
numeric_columns = dataset.select_dtypes(include=np.number)
correlation_matrix = numeric_columns.corr()
```

```python

# Assuming dataset is your DataFrame
numeric_columns = dataset.select_dtypes(include=np.number)
correlation_matrix = numeric_columns.corr()

# Visualize the correlation matrix using a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
```

```python

# Selecting only numeric columns
numeric_columns = dataset.select_dtypes(include=np.number)

# Computing correlation matrix
correlation_matrix = numeric_columns.corr()

# Visualizing the correlation matrix using a heatmap
sns.heatmap(correlation_matrix, cmap='Reds', annot=True)
```

```python
import seaborn as sns

# Selecting only numeric columns
numeric_columns = dataset.select_dtypes(include=np.number)

# Computing correlation matrix
correlation_matrix = numeric_columns.corr()

# Visualizing the correlation matrix using a heatmap
sns.heatmap(correlation_matrix, cmap='Blues', annot=True)
```
